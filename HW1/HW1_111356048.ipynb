Q1. Louisa's muffins

- Maximize: 
  - (1.6 * TP + 1.4 * TC + 1.9 * TN + 1.2 * TT)
- Constraint:
  - TP + TC + TN + TT <=1200
  - 310 <= TP <= 434
  - 245 <= TC <= 343
  - 255 <= TN <= 357
  - 190 <= TT <= 266


```python
%pip install gurobipy>=10
import gurobipy as grb
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.stats
```

```python
mMuf = grb.Model('Muffins')

I = 4

pieces_var={}
for i in range(I):
  pieces_var[i]= mMuf.addVar(vtype=grb.GRB.CONTINUOUS, name='x'+str(i))

pieces_var
```

```python
Demand = [620, 490, 510, 380]
Profit = [1.6, 1.4, 1.9, 1.2]
```

```python
mMuf.addConstr((pieces_var[0] +pieces_var[1]+pieces_var[2]+pieces_var[3]) <= 1200)
for i in range(I):
  mMuf.addConstr(pieces_var[i] >= Demand[i] * 0.5)
  mMuf.addConstr(pieces_var[i] <= Demand[i] * 0.7)
```

```python
objectiveMuf = grb.quicksum(pieces_var[i]* Profit[i] for i in range(I))

mMuf.setObjective(objectiveMuf, grb.GRB.MAXIMIZE)
mMuf.update()
mMuf.optimize()
```

```python
print(mMuf.Status == grb.GRB.OPTIMAL)
print(mMuf.display())
```

```python
for v in mMuf.getVars():
  print(v.VarName, v.X)
```

Ｑ2. Cattle feed
- Minimize: 
  - (200 * oats + 150 * corn + 100 * alfalfa + 75 * peanut hulls)
- Constraint:
  - protein + fiber + fat = 1 
  - protein > 60%
  - fiber > 60%
  - fat < 60%

```python
model = grb.Model("cattle")

I=4
x = model.addVars([i for i in range(I)], vtype = grb.GRB.CONTINUOUS, name = "x")
```

```python
cost = [200, 150, 100, 75]
protein = [0.6, 0.8, 0.55, 0.4]
fat= [0.5, 0.7, 0.4, 1]
fiber = [0.9, 0.3, 0.6, 0.8]
```

```python
model.addConstr(grb.quicksum(x[i] for i in range(I)) == 1)
model.addConstr(grb.quicksum(x[i] * protein[i] for i in range(I)) >=0.6)
model.addConstr(grb.quicksum(x[i] * fiber[i] for i in range(I)) >=0.6)
model.addConstr(grb.quicksum(x[i] * fat[i] for i in range(I)) <=0.6)
```

```python
obj = (grb.quicksum(x[i] * cost[i] for i in range(I)))
model.setObjective(obj, grb.GRB.MINIMIZE)
model.update()
model.optimize()
```

```python
print(model.Status == grb.GRB.OPTIMAL)
print(model.display())
```

```python
for v in model.getVars():
  print(v.VarName, v.X)
```

```python
optobj = model.getObjective()
print(optobj.getValue())
```

Q3. 4 gasolines
- Minimize: 
  - (48 * gas1 + 43 * gas2 + 58 * gas3 + 46 * gas4)
- Constraint:
  - 80 < first < 90 
  - 270 < sec < 280
  - gas1 + gas2 + gas3 + gas4 = 1




```python
mGas = grb.Model('gasoline')
I = 4
J = 2
x = mGas.addVars([i for i in range(I)], vtype = grb.GRB.CONTINUOUS, name = "x")
x
```

```python
quality = [[99, 210], [70, 335], [78, 280], [91, 265]]
cost = [48, 43, 58, 46]
```

```python
mGas.addConstr(grb.quicksum(x) == 1)

mGas.addConstr(grb.quicksum(x[i] * quality[i][0] for i in range(I)) >= 80)
mGas.addConstr(grb.quicksum(x[i] * quality[i][0] for i in range(I)) <= 90)
mGas.addConstr(grb.quicksum(x[i] * quality[i][1] for i in range(I)) >= 270)
mGas.addConstr(grb.quicksum(x[i] * quality[i][1] for i in range(I)) <= 280)
```

```python
obj = grb.quicksum(x[i] * cost[i] for i in range(I))
mGas.setObjective(obj,grb.GRB.MINIMIZE)
mGas.update()
mGas.optimize()
```

```python
print(mGas.Status == grb.GRB.OPTIMAL)
print(mGas.display())
```

```python
for v in mGas.getVars():
  print(v.VarName, v.X)
```

```python
#Print optimal objective value
obj=mGas.getObjective()
print(obj.getValue())
```
