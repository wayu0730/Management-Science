```python
%pip install gurobipy>=10
import gurobipy as grb
import numpy as np
import math
```

```python
model_2a = grb.Model('m2a')
# I: Number of Months
# J: Number of Servers
I = 5
J = 4
x = model_2a.addVars([(i, j) for i in range(I) for j in range(J)], vtype = grb.GRB.INTEGER, name = "x")
x
```

```python
numOfEmp = np.array([60, 200, 30, 75])
costOfServer = np.array([[2500, 5000, 9000, 18750],[2500, 5000, 9000, 18750],[2500, 5000, 10000, 25000],[2500, 5000, 10000, 25000],[2500, 5000, 10000, 25000]])
limitOfServer = np.array([30,80,200,2000])
```

## Constraint
1. Manufacturing dept. requires at least one of the three powerful servers.
2. The number of employees server support in the first 2 months must be at least 60.
3. The number of employees server support in the first 3 months must be at least 260
4. The number of employees server support in the first 4 months must be at least 290
5. The number of employees server support in the first 5 months must be at least 365
6. The budget for the first 2 months is less than $9500

```python
model_2a.addConstr(x[2,1] + x[2,2] + x[2,3] >=1)
model_2a.addConstr(grb.quicksum(limitOfServer[j] * x [i, j] for i in range(2) for j in range(J)) >= 60)
model_2a.addConstr(grb.quicksum(limitOfServer[j] * x [i, j] for i in range(3) for j in range(J)) >= 260)
model_2a.addConstr(grb.quicksum(limitOfServer[j] * x [i, j] for i in range(4) for j in range(J)) >= 290)
model_2a.addConstr(grb.quicksum(limitOfServer[j] * x [i, j] for i in range(5) for j in range(J)) >= 365)
model_2a.addConstr(grb.quicksum(costOfServer[i, j] * x [i, j] for i in range(2) for j in range(J)) <= 9500)
```

## Objective
* Minimize the total cost of 5 months in 4 servers.

```python
obj_2a = grb.quicksum(costOfServer[i, j] * x [i, j] for i in range(I) for j in range(J))
model_2a.setObjective(obj_2a, grb.GRB.MINIMIZE)
model_2a.update()
model_2a.optimize()
```

```python
print(model_2a.Status == grb.GRB.OPTIMAL)
print(model_2a.display())
```

# **3-2 b.**

```python
%pip install pulp
from pulp import *

# Define the decision variables
types = ['A', 'B', 'C']
months = [1, 2, 3, 4, 5]
servers = LpVariable.dicts("servers", [(t, m) for t in types for m in months], lowBound=0, cat='Integer')

# Define the problem
prob = LpProblem("Server Purchase Problem", LpMinimize)

# Define the constraints
departments = {'sales': [50, 50, 50, 50, 50],
               'manufacturing': [0, 100, 150, 150, 150],
               'warehouse': [0, 0, 50, 100, 100],
               'marketing': [0, 0, 0, 50, 100]}
for dept, reqs in departments.items():
    for m in months:
        prob += lpSum(servers[(t, m)] for t in types) >= reqs[m-1]

for t in types:
    for m in months:
        if t == 'A':
            prob += servers[(t, m)] <= 50
        elif t == 'B':
            prob += servers[(t, m)] <= 100
        elif t == 'C':
            prob += servers[(t, m)] <= 150

# Define the objective function
costs = {'A': 5000, 'B': 10000, 'C': 15000}
prob += lpSum(servers[(t, m)] * costs[t] for t in types for m in months)

# Solve the problem
prob.solve()

# Print the results
for t in types:
    for m in months:
        print(f"Month {m}: Purchase {value(servers[(t, m)])} server(s) of type {t}")
print(f"Total cost: {value(prob.objective)}")
```
